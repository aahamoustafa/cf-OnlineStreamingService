# Exploring the Data # 

```python
import pandas as pd 
import os 
import seaborn as sns
import matplotlib.pyplot as plt 
path = r'/Users/aahamoustafa/Desktop/Data Projects/UFood Marketing Analysis 12-2024'
```

```python
food = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'u_food_marketing_cleaned.csv'), index_col = False)

pd.set_option('display.max.columns', 50)
pd.set_option('display.max.rows', 2300)
```

```python
## dropping unnamed column 
food = food.iloc[:, 1:]
```

```python
food.head()
```

```python
## correlation exploration 

food_numeric = food.select_dtypes(include=['float64', 'int64'])

food_numeric.corr(method = 'pearson')['AcceptedCmpOverall'].sort_values(ascending = False)
```

```python
## segmenting customers by age and creating age groups 

food['Age'].sort_values()

age_groups = [(23,30),(31,40),(41,50),(51,60),(61,70),(71,85)]

def assign_age_group(Age): 
    for age_range in age_groups: 
        if age_range[0] <= Age <= age_range[1]:
            return f"{age_range[0]} - {age_range[1]}"
    return("Unknown")

food['Age group'] = food['Age'].apply(assign_age_group)
```

```python
## visualizing campgains against age 
age_order = ['23 - 30','31 - 40','41 - 50','51 - 60','61 - 70','71 - 85']
sns.pointplot( data = food, x = 'Age group', y = 'AcceptedCmpOverall', order = age_order )
```

#### showing the overall acceptance rate for each of the age groups. the lines are confidence intervals ( really large line then its not super confident in these numbers) --> maybe there is a small sample size in the 23-30 and 71-85 age groups

```python
## to improve the skew in the data, we need to know how many people are actually in each age group 

counts = food['Age group'].value_counts()
```

#### this confirms the suspicion and thats why the confidence intervals were much higher in the middle age groups 

```python
## creating a percentage age dataframe
percentage = counts /food.shape[0]
percent_food = percentage.reset_index()
percent_food.columns = ['age_group','percentage']
percent_food.sort_values('age_group', inplace = True)
```

```python
percent_food
```

```python
sns.barplot(data = percent_food, x = 'age_group', y = 'percentage')
plt.title ('Percentage of Accepted Campgains per Age Group')
plt.show()
```

#### Age Segmentation - Our core audience for accepting campagins right now is 31 - 70 , but how much does each group spend? This can change our core audience recommendation

```python
grouped_amount_spent = food.groupby('Age group')['MntTotal'].sum().reset_index()


sns.barplot(x = 'Age group', y = 'MntTotal', data = grouped_amount_spent)
plt.title('Amount Spent Per Age Group')
plt.show()
```

#### Our recommendation does not change as the spending power is the same as the percentage of accepted campaigns. 23-30, 71 and up accept at higher rates.

```python
## How are Purchases Divided based on Web, Catalog, & Store Purchases?

sum_food = pd.DataFrame(food[['NumWebPurchases','NumCatalogPurchases','NumStorePurchases']].sum(), columns = ['Sums']).reset_index()
sum_food.rename(columns = {'index': 'Type_of_Purchase'}, inplace = True)
sns.barplot(x = 'Type_of_Purchase', y = 'Sums', data = sum_food)
```

```python
## what about the sums of those who accepted campaigns?

acct_camp = food[food['AcceptedCmpOverall'] != 0]

sum_food = pd.DataFrame(acct_camp[['NumWebPurchases','NumCatalogPurchases','NumStorePurchases']].sum(), columns = ['Sums']).reset_index()
sum_food.rename(columns = {'index': 'Type_of_Purchase'}, inplace = True)
sns.barplot(x = 'Type_of_Purchase', y = 'Sums', data = sum_food)
plt.show()
```

#### 2 Directions: Either Boost up the higher percentage catalog customers. OR focus In Store/Web because they have more traffic

```python
#### what about spending based on total children who accept the campaings?

sns.regplot(x = 'Total Children', y = 'AcceptedCmpOverall', data = food) 
```

#### Less kids = more likely to accept campaign due to the negative slope

```python
#### what about spending based on education level

sns.regplot(x = 'Education Status', y = 'AcceptedCmpOverall', data = food) 
```

#### As Education level gets higher, spending gets higher. But Education isn't really significant in our segmentation

```python
#### what about spending based on marital status & accepting campaings?

acct_camp = food[food['AcceptedCmpOverall'] != 0]

marital_spending = acct_camp.groupby('Marital_Status_Str')['MntTotal'].sum().reset_index()
sns.barplot(x = 'Marital_Status_Str', y = 'MntTotal', data = marital_spending) 
```

#### Married Single & Together are spending a lot more money than widow and divorced.

## OVERALL FINDINGS

### 1. Age - 30-70 were spending more money but less likely to accept campaigns. 

### 2. Catalog Purchases more likely to accept campaigns, but in person spend more. Recommend a split between them ( 40% catalog, 30% store, 30% web ) 

### 3. Focus on people with no kids( or less kids)

### 4. Education - no impact - dont target any group

### 5. Marital Status - does not play a big part. Married,Single,Together spend more money 

## Recommendation

### 1. For Profit - Target Middle aged people, high earner, with no kids. Target on different platforms with split above

### 2. For New Users to Spend Money ( accept more campaigns)  - Focus on 21-30, 70 and up - who statistcally accepted campaigns at a higher rate
